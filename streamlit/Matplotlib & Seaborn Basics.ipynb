# Minimal Line plot

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3],[4,5,6],label="Line One")
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.grid(True)
plt.show()
```

# Scatter Plot

```python
plt.scatter([1,2,3],[4,5,6])
plt.show()
```

# Bar chart

```python
plt.bar(['A','B','C'],[5,10,15])
plt.show()
```

# Histogram

```python
plt.hist([1,2,2,3,3,3,4,4,4,4])
plt.show()
```

# Subplots

```python
fig,axs=plt.subplots(1,2) # Show two plots side by side
axs[0].plot([1,2,3],[4,5,6])
axs[1].bar(['A','B','C'],[10,15,20])
plt.show()
```

# Seaborn

```python
import seaborn as sns
tips=sns.load_dataset('tips')
print(tips)
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
sns.set_style('darkgrid')
plt.title("Dark Grid")
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.boxplot(x='day',y='tip',data=tips)
plt.show()
```

```python
corr=tips.select_dtypes(include='number').corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```
